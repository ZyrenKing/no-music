```python
#@title 🎵 Razi's Colab Audio Isolation System 🎵

import subprocess
import sys
import torch
import shutil
import os
from google.colab import drive

# --- Colab Form Fields ---
#@markdown Enter the original audio file name:
input_mp3 = "out.mp3" #@param {type:"string"}

#@markdown Select Processing Mode:
processing_mode = "2. Balanced (High Quality, Good Speed)" #@param ["1. Maximum Speed (Lower Quality)", "2. Balanced (High Quality, Good Speed)", "3. Maximum Quality (Slower)"]

def process_audio():
    # ربط قوقل درايف تلقائياً
    print("🔗 جاري ربط Google Drive...")
    drive.mount('/content/drive')

    # Determine shifts based on selection
    if processing_mode.startswith("1"):
        shifts = 0
    elif processing_mode.startswith("3"):
        shifts = 2
    else:
        shifts = 1 # Default Balanced

    device = "cuda" if torch.cuda.is_available() else "cpu"

    command = f'python -m demucs -n htdemucs_ft -d {device} --mp3 --two-stems=vocals --shifts {shifts} -j 2 "{input_mp3}"'

    print(f"\n⚙️ Processing '{input_mp3}' with {device.upper()} (Shifts: {shifts}), please wait...\n")

    try:
        process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

        for char in iter(lambda: process.stdout.read(1), ''):
            sys.stdout.write(char)
            sys.stdout.flush()

        process.wait()

        if process.returncode == 0:
            print("\n\n✅ Isolation successful! Files are in the 'separated' folder.")

            # استخراج اسم الملف بدون الامتداد لمعرفة مسار المجلد الذي أنشأته الأداة
            base_name = os.path.splitext(input_mp3)[0]
            source_file = f"/content/separated/htdemucs_ft/{base_name}/vocals.mp3"

            # مسار الحفظ في قوقل درايف (سيتم حفظه باسم الملف الأصلي + كلمة clean)
            destination = f"/content/drive/MyDrive/{base_name}_clean.mp3"

            print("⏳ جاري نقل الملف إلى Google Drive...")
            shutil.copy(source_file, destination)
            print(f"✅ تم حفظ الملف بنجاح! تجده في درايف الخاص بك باسم: {base_name}_clean.mp3")

        else:
            print(f"\n\n❌ Process failed with error code {process.returncode}")

    except Exception as e:
        print(f"\n\n❌ An error occurred: {e}")

if __name__ == "__main__":
    if not input_mp3:
        print("❌ Please enter a valid audio file name in the form above.")
    else:
        process_audio()
```
